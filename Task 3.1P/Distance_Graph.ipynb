```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
df = pd.read_csv("Distance.csv")
```

```python
df
```

```python
# Ensure the 'timestamp' column is in datetime format
df["time"] = pd.to_datetime(df["time"])
df
```

```python
# Set the 'timestamp' column as the index
df.set_index("time", inplace=True)

# Resample the data to 10-second intervals
df_resampled = df.resample("20S").mean()

# Plot the resampled data
plt.figure(figsize=(10, 6))
plt.plot(df_resampled.index, df_resampled["value"], marker="o", linestyle="-", color="b", label="Line Plot")
plt.title("Line Plot with 20-Second Intervals")
plt.xlabel("Timestamp")
plt.ylabel("Value")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
