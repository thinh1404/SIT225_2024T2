```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Plotting X

```python
X_df = pd.read_csv("iPhone Thing-Accelerometer_X.csv")
Y_df = pd.read_csv("iPhone Thing-Accelerometer_Y.csv")
Z_df = pd.read_csv("iPhone Thing-Accelerometer_Z.csv")
X_df['time'] = pd.to_datetime(X_df['time'])
Y_df['time'] = pd.to_datetime(Y_df['time'])
Z_df['time'] = pd.to_datetime(Z_df['time'])
```

```python
# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 12))

# Plot X Accelerometer data
sns.lineplot(data=X_df, x='time', y='value', ax=axs[0], color='red')
axs[0].set_title('X Accelerometer Line Graph')

# Plot Y Accelerometer data
sns.lineplot(data=Y_df, x='time', y='value', ax=axs[1], color='green')
axs[1].set_title('Y Accelerometer Line Graph')

# Plot Z Accelerometer data
sns.lineplot(data=Z_df, x='time', y='value', ax=axs[2], color='blue')
axs[2].set_title('Z Accelerometer Line Graph')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a single plot
plt.figure(figsize=(12, 6))

# Plot X Accelerometer data
sns.lineplot(data=X_df, x='time', y='value', label='X Accelerometer', color='red')

# Plot Y Accelerometer data
sns.lineplot(data=Y_df, x='time', y='value', label='Y Accelerometer', color='green')

# Plot Z Accelerometer data
sns.lineplot(data=Z_df, x='time', y='value', label='Z Accelerometer', color='blue')

# Set the title and labels
plt.title('Accelerometer Data for X, Y, Z Axes')
plt.xlabel('Time')
plt.ylabel('Value')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
